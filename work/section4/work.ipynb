```python
import re
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import time
```

```python
# 回答
url = "https://www.python.org/"
res = requests.get(url)
# アクセスできるか確認
# print(res.status_code)

soup = BeautifulSoup(res.content, "lxml") 
# Latest Newsがあるブロックを取得
post = soup.find("div", class_="medium-widget blog-widget")

d_list = []
for i, li in enumerate(post.find_all("li")):
    title = li.find("a").text
    date = li.find("time").text
    url = li.find("a").get("href")
    # 辞書に格納
    dict = {
        "title": title,
        "url" : url,
        "date": date
    }
    d_list.append(dict)

df = pd.DataFrame(d_list)
# csvファイルに出力
# df.to_csv("python_web_posts.csv", index=None, encoding="utf-8-sig")
df.to_excel("python_web_posts.xlsx", index=None)
```

```python
# 回答
url = "https://www.python.org/"
res = requests.get(url)
# アクセスできるか確認
# print(res.status_code)

soup = BeautifulSoup(res.content, "lxml") 
# Latest Newsがあるブロックを取得
post = soup.find("div", class_="medium-widget blog-widget")

d_list = []
for i, li in enumerate(post.find_all("li")):
    title = li.find("a").text
    date = li.find("time").text
    post_url = li.find("a").get("href")

    # 休止状態を入れる
    time.sleep(3)

    post_url = requests.get(post_url) # for文で記事のURLにアクセス -> 休止状態を入れる
    post_soup = BeautifulSoup(post_url.content, "lxml")
    # 記事のタイトル(h2)を取得
    post_h2 = [h2.text for h2 in post_soup.find_all("h2")]

    # 辞書に格納
    dict = {
        "title": title,
        "url" : url,
        "date": date,
        "post_h2": post_h2
    }
    d_list.append(dict)

df = pd.DataFrame(d_list)
# csvファイルに出力
# df.to_csv("python_web_posts.csv", index=None, encoding="utf-8-sig")
# df.to_excel("python_web_posts.xlsx", index=None)
```
