```python
# 回答例
base_url = "https://employment.en-japan.com" 
url = "https://employment.en-japan.com/search/search_list/?occupation=101000_102500_103000_103500_104500_105000_105500_109000&areaid=23_24_21_50"

res = requests.get(url)
res.raise_for_status()
time.sleep(2)

soup = BeautifulSoup(res.text, "lxml")
# 企業が乗っているブロックを取得
# jobs = soup.find_all("div", class_="unitBase") # 全ての企業情報を取得
# print(len(jobs)) # 本来は50件になるはず -> 取得する場所が違う

# 企業の名前を取得する
jobs = soup.find_all("div", class_="jobNameArea")
d_list = []
count = 0
for job in jobs[:30]:
    count += 1
    company_name = job.find("span", class_="company").text
    # list.append(company_name)
    page_url = base_url + job.find("a").get("href")

    page_res = requests.get(page_url)
    page_res.raise_for_status()
    time.sleep(3)

    company_url = None

    page_soup = BeautifulSoup(page_res.content, "lxml")
    company_info_block = []
    if "PK" in page_url:
        company_summaries = page_soup.find_all("div", class_="descArticleUnit dataCompanyInfoSummary")
        for company_summary in company_summaries:
            if "会社概要" in company_summary:
                company_info_block.append(company_summary)

            company_summary = company_info_block[0]
        # print(company_summary)
        
        for table_row in company_summary.find_all("tr"):
            if table_row.find("th").text == "企業ホームページ":
                company_url = table_row.find("td").find("a").text
                # print(company_url)

    elif "fromSearch" in page_url:
        detail_page_url = page_soup.find("iframe", id="recruitFrame").get("src")
        time.sleep(3)
        detail_page_res = requests.get(detail_page_url)
        detail_page_res.raise_for_status()
        detail_page_soup = BeautifulSoup(detail_page_res.content, "lxml")
        company_summary = detail_page_soup.find("table", class_="companyTable")
        # print(company_summary)
        # print(company_summary)
        for table_row in company_summary.find_all("tr"):
            if table_row.find("th").text == "企業WEBサイト":
                company_url = table_row.find("td").find("a").get("href")
                # print(company_url)

    d_list.append({
        "company_name": company_name,
        "company_url": company_url
    })
    
    print(d_list[-1])
```

```python
from time import sleep
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = 'https://employment.en-japan.com'
url = 'https://employment.en-japan.com/wish/search_list/?companytype=0&worktype=0&areaid=23_24_21_50&occupation=101000_102500_103000_103500_104500_105000_105500_109000&indexNoWishArea=0&sort=wish&pagenum={}'
max_page_index = 3
d_list = []

for i in range(max_page_index):
    print('d_list:', len(d_list))
    access_url = url.format(i+1)
    sleep(3)
    
    r = requests.get(access_url)
    r.raise_for_status()
    
    soup = BeautifulSoup(r.content, 'lxml')
    jobs = soup.find_all('div', class_='jobNameArea')
    
    print(access_url, len(jobs))
    
    for job in jobs:
        company_name = job.find('span', class_='company').text
        page_url = base_url + job.find('a').get('href')
        
        if 'fromSearch' in page_url:
            path = page_url.replace(base_url+'/desc_eng_', '')
            e_index = path.rfind('/')
            company_id = path[:e_index]
            page_url = f'https://en-gage.net/recruit/?getFromEmploy={company_id}'
        
        sleep(3)
        
        page_r = requests.get(page_url)
        page_r.raise_for_status()
        
        page_soup = BeautifulSoup(page_r.content, 'lxml')
        company_url = None
        
        if 'PK' in page_url:
            company_info_block = [h2 for h2 in page_soup.find_all('h2', class_='text') if '会社概要' in h2.text][0]
            company_summary = company_info_block.parent.parent
            
            for table_row in company_summary.find_all('tr'):
                if table_row.find('th').text == '企業ホームページ':
                    company_url = table_row.find('td').find('a').text

        elif 'getFromEmploy' in page_url:
            company_summary = page_soup.find('table', class_='companyTable')
            
            for table_row in company_summary.find_all('tr'):
                if table_row.find('th').text == '企業WEBサイト':
                    company_url = table_row.find('td').find('a').get('href')
                else:
                    raise Exception("Table row structure does not match expected pattern")

        d_list.append({
            'company_name': company_name,
            'company_url': company_url
        })
        print(d_list[-1])

# データフレームにしてCSVに保存
df = pd.DataFrame(d_list)
df.to_csv('enjapan_company_list.csv', index=False, encoding='utf-8-sig')
```
